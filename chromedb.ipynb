```python
import chromadb
import pandas as pd

# Bỏ giới hạn hiển thị pandas
pd.set_option('display.max_rows', None)        # hiện hết rows
pd.set_option('display.max_columns', None)     # hiện hết columns
pd.set_option('display.max_colwidth', None)    # hiện hết text trong cell
pd.set_option('display.width', None)           # không wrap dòng
```

```python
client = chromadb.PersistentClient(path="chroma_db")
col = client.get_collection("rag_lab")

results = col.get(include=["documents", "metadatas"])

df = pd.DataFrame(results["metadatas"])
df["text"] = results["documents"]
df["chunk_id"] = results["ids"]

# Sắp xếp cột cho dễ đọc
df = df[["chunk_id", "source", "department", "section", "effective_date", "access", "text"]]
df
```

```python
results_with_vec = col.get(include=["documents", "metadatas", "embeddings"])

df_vec = pd.DataFrame(results_with_vec["metadatas"])
df_vec["text"] = results_with_vec["documents"]
df_vec["chunk_id"] = results_with_vec["ids"]
df_vec["embedding_dim"] = [len(e) for e in results_with_vec["embeddings"]]
df_vec["embedding_preview"] = [str(e[:5]) + "..." for e in results_with_vec["embeddings"]]

# ✅ Fix: lưu từng vector riêng lẻ dưới dạng object
import numpy as np
df_vec["embedding_full"] = pd.Series(
    [np.array(e) for e in results_with_vec["embeddings"]],
    dtype=object
)

df_vec[["chunk_id", "source", "section", "text", "embedding_dim", "embedding_preview"]]
```
